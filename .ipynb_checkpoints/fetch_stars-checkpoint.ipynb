#### Import Statements

```python
# Data structures
import numpy as np 
import pandas as pds

# Plotting
from matplotlib import pyplot as plt

# Web-scraping
import requests 
from io import StringIO 
```

#### Background Information

There are a few things to consider before fetching the data, which comes from [this location](http://skyserver.sdss.org/dr14/en/home.aspx).  We are primarily focusing on the `photoObj` properties, though certainly you could [look elsewhere](http://skyserver.sdss.org/dr7/en/help/docs/tabledesc.asp).

First, this is all stored in a database object.  As such, it is necessary to query the data using the necessary keywords.  That is provided below.  

Second, it is a large dataset.  As such, it would be better to instead fetch a portion of the entire set, as opposed to everything.  We do this by the right ascension (an angle in degrees) and the declination (an angle in degrees, also).  By limiting it such that

$x_{lower} < x < x_{upper}$

where $x$ can either be right ascension or declination, we can instead retrieve a patch of the sky.  However, one thing to note here is that a negative angle would also correspond to $360 + x$.  Choose your bounds carefully.  

#### Fetching the Data

```python
def fetch_patch(ra, dec): 
    
    # Note:
    # ra -> right ascension
    # dec -> declination
    
    ra_lower, ra_upper = ra
    dec_lower, dec_upper = dec
    
    # ra_lower -> lower right ascension bound
    # ra_upper -> upper right ascension bound
    # dec_lower -> lower declination bound
    # dec_upper -> upper declination bound
    
    # query -> the string being sent to the SkyServer database
    # You can find what all these variables mean on the site, 
    # You can also include more variables if desired.
    
    query = '''SELECT
       p.objid, p.ra, p.dec, p.u, p.g, p.r, p.i, p.z, p.cx, p.cy, p.cz,
       s.class, s.z as redshift
    FROM PhotoObj AS p
       JOIN SpecObj AS s ON s.bestobjid = p.objid
    WHERE 
       p.ra between {} and {} and
       p.dec between {} and {}
    '''.format(ra_lower, ra_upper, dec_lower, dec_upper)
    
    # The location of the site.
    url = 'http://skyserver.sdss.org/dr14/SkyServerWS/SearchTools/SqlSearch?cmd={}&format=csv'.format(query)
    
    # Making a request on the data.
    r = requests.get(url)
    
    # Check for a succesful result.
    if r.status_code == 200:
        
        # This creates a csv file from the text.
        csv = StringIO(r.text)
        
        # Making a dataframe, since operations on that are easier.
        df = pds.read_csv(csv, skiprows = 1)
        
        return df
```

```python
# Start with something small, say, 1 degree across both.
ra = (0, 1)
dec = (0, 1)

patch = fetch_patch(ra, dec)
```

```python
# Show the top few rows of the dataframe:
patch.head()
```

Here is a list of variables and a brief description:

* `objid`: a unique identifier for that particular celestial object.  Think of these like names, but purely integers.
* `ra`: right ascension
* `dec`: declination
* `u`, `g`, `r`, `i`, `z`: elements of the [Photometric System](https://en.wikipedia.org/wiki/Photometric_system) with unknown units (need to be researched)
* `cx`, `cy`, `cz`: unit vectors for the spatial location of that object.  Note that their magnitude is one, so it cannot be used as an absolute distance.
* `class`: the type of object.  
* `redshift`: the value for redshift or a metric of distance away from the observatory with unknown units (also needs to be researched)

These in all in some way definitely serve a use.  Namely, we can color code according to the different classes.  An example of that:

```python
# Note that you can get a column by indexing the dataframe like a dictionary, by using the column name as a key.
classes = patch['class']

# classes.unique returns all of the unique elements in that set.
print(classes.unique())
```

We can then, say, allow galaxies to be colored a deep blue, stars a gold, an the [quasi-stellar objects (QSO)](https://en.wikipedia.org/wiki/Quasar) a grey color:

```python
# We create a color dictionary so we can index this instead of if-else statements.
class_colors = {'GALAXY': 'navy', 'STAR': 'gold', 'QSO': 'dimgray'}

# We can also create markers for scatterplots:
class_markers = {'GALAXY': 's', 'STAR': 'o', 'QSO': '^'}
```

Now, we can plot according to their class:

```python
fig, ax = plt.subplots(1, 1)

for cls in classes.unique():
    cls_patch = patch[patch['class'] == cls]
    
    ax.scatter(cls_patch['ra'], cls_patch['dec'], 
               c = class_colors[cls], marker = class_markers[cls],
               alpha = 0.5, s = 25, label = cls)
    
ax.set_ylabel('declination, degrees', fontsize = 14)
ax.set_xlabel('right ascension, degrees', fontsize = 14)

fig.legend(bbox_to_anchor = (1.15, 0.925), fontsize = 12, frameon = False);
```

So, this is cool and all...but what are we trying to do? Basically, there is no goal in mind.  The point of research in general is to explore rather basic topics and try to find something interesting.  This can be achieved in one of two ways:  by banging your head against the wall and hoping things work, or by doing background research and trying to first recreate and then expand on that work.  

But what is interesting? Interesting from a data science perspective is that the data shows some kind of behavior, especially when groomed such that classification can be observed.  Take this for example:

```python
fig, ax = plt.subplots(1, 1)

for cls in classes.unique():
    cls_patch = patch[patch['class'] == cls]
    
    ax.scatter(cls_patch['u'], cls_patch['z'], 
               c = class_colors[cls], marker = class_markers[cls],
               alpha = 0.5, s = 25, label = cls)
    
ax.set_ylabel('$u$', fontsize = 14)
ax.set_xlabel('$g$', fontsize = 14)

fig.legend(bbox_to_anchor = (1.15, 0.925), fontsize = 12, frameon = False);
```

Here, it doesn't appear that there's much overlap when comparing $u$ and $g$, separated by object class.  This is interesting.  Why is this interesting? Clearly something explainable is going on, but we don't know what.  And that's exactly what the purpose of this is.  Perhaps the linearity can be explained away by the inherent overlap of `ugriz`, since they're bands of light, but what about the class distinction? This is where the background research comes into play.
